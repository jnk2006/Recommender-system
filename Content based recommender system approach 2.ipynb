### Merging columns

<p>Merge columns (artistid, artistname, tagid) and store them in a new dataframe artists.</p>

```python
#rename the column id in artists_df to artistID to make easier to join the dataframes.
artists_df.rename(columns = {'id': 'artistID'}, inplace = True)

#initially merge artists_df and user_taggedartists_df as they both contain artistID
artists = pd.merge(artists_df, user_taggedartists_df, on='artistID', how='inner')

#Then merge artists and tags_df
artists = pd.merge(artists, tags_df, on = 'tagID', how = 'inner')
print(artists.shape)
artists.head()
```

Drop unnecessary columns.

```python
artists.drop(['url', 'pictureURL', 'userID', 'day', 'month', 'year', 'tagValue'], inplace = True, axis = 1)
artists.head(10)
```

Since the name column contains non-alphanumeric values, initally replace those values with NaN and then drop rows that contain NaN values.

```python
artists['name'].replace(r'\W', np.nan, regex=True, inplace=True)
artists.dropna(inplace = True)
```

<p>This encoding is needed for feeding categorical data. In this case, we store every different tag value in columns that contain either 1 or 0. 1 shows that an artist has that genre and 0 shows that it doesn't.</p>

<p>Let's also store this dataframe in another variable since tags won't be important for our first recommendation system.</p>
