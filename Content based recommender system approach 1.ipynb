### Merging columns

# Content-Based recommendation system

Now, let's take a look at how to implement __Content-Based__ or __Item-Item recommendation systems__. This technique attempts to figure out what a user's favourite aspects of an item is, and then recommends items that present those aspects. In our case, we're going to try to figure out the input's favorite tags from the artists and ratings given.

Let's begin by creating an input user to recommend artists to:

Notice: To add more artists, simply increase the amount of elements in the __userInput__. Feel free to add more in!

```python
userInput = [
            {'name':'OneRepublic', 'weight':4576},
            {'name':'MGMT', 'weight':484},
            {'name':'Madonna', 'weight':8473},
            {'name':"Shakira", 'weight':5875},
            {'name':'Rihanna', 'weight':4750},
            {'name':'Adele', 'weight':8887}
         ] 
inputArtists = pd.DataFrame(userInput)
inputArtists
```

#### Add artistId to input user
With the input complete, let's extract the input artist's ID's from the artists dataframe and add them into it.

We can achieve this by first filtering out the rows that contain the input artist's name and then merging this subset with the input dataframe. We also drop unnecessary columns for the input to save memory space.

```python
#Filtering out the artists by name
inputId = artists_df[artists_df['name'].isin(inputArtists['name'].tolist())]

#Then merging it so we can get the artistId. It's implicitly merging it by name.
inputArtists = pd.merge(inputId, inputArtists)

#Dropping information we won't use from the input dataframe
inputArtists = inputArtists.drop('url', 1).drop('pictureURL', 1)
#Final input dataframe
inputArtists
```

We're going to start by learning the input's preferences, so let's get the subset of artists that the input has listened to from the Dataframe containing tags defined with binary values.

```python
#Filtering out the artists from the input
userArtists = artistsWithTags_df[artistsWithTags_df['artistID'].isin(inputArtists['artistID'].tolist())]
userArtists
```

```python
userArtists = userArtists.groupby(['artistID', 'name'], as_index = False).mean()
userArtists
```

We'll only need the actual tags table, so let's clean this up a bit by resetting the index and dropping the artistID and name column.

```python
#Resetting the index to avoid future issues
userArtists = userArtists.reset_index(drop=True)

#Dropping unnecessary issues due to save memory and to avoid issues
userTagTable = userArtists.drop('artistID', 1).drop('name', 1)
userTagTable
```

Now we're ready to start learning the input's preferences!

To do this, we're going to turn each tag into weights. We can do this by using the input's weights and multiplying them into the input's tag table and then summing up the resulting table by column. This operation is actually a dot product between a matrix and a vector, so we can simply accomplish by calling Pandas's "dot" function.

```python
inputArtists['weight']
```

```python
#Dot produt to get weights
userProfile = userTagTable.transpose().dot(inputArtists['weight'])
#The user profile
userProfile
```

Now, we have the weights for every of the user's preferences. This is known as the User Profile. Using this, we can recommend artists that satisfy the user's preferences.

Let's start by extracting the tag table from the original dataframe:

```python
#Now let's get the genres of every movie in our original dataframe
tagTable = artistsWithTags_df.set_index(artistsWithTags_df['artistID'])

tagTable.head()
```

```python
#And drop the unnecessary information
tagTable = tagTable.drop('artistID', 1).drop('name', 1)
tagTable.head()
```

```python
tagTable.shape
```

With the input's profile and the complete list of artists and their tags in hand, we're going to take the weighted average of every artist based on the input profile and recommend the top twenty artists that most satisfy it.

```python
#Multiply the genres by the weights and then take the weighted average
recommendationTable_df = ((tagTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
#Sort our recommendations in descending order
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
#Just a peek at the values
recommendationTable_df.head()
```

Now here's the recommendation table!

```python
#The final recommendation table
final_recomm = artists_df.loc[artists_df['artistID'].isin(recommendationTable_df.head(20).keys())]
```
